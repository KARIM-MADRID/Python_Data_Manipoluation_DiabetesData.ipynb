```python
from sklearn import preprocessing

# This method provides the normalization of all columns except label column
def get_normalizatoin(df):
    y = df['Outcome']  
    x = (df.drop('Outcome',axis = 1)).values  #Dropping label colun as we do not need to normalize it
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled)
    df['Outcome'] = y # Appedning a label column back to dataset
    return df

# This method provides the standardization of all columns except label column
def get_standardization(df):
    y = df['Outcome']
    df = df.drop('Outcome', axis=1)
    df = (df-df.mean())/df.std()  # Standardization = (data - mean)/standard deviation
    df['Outcome'] = y  # Appedning a label column back to dataset
    return df
```

```python
from sklearn.model_selection import train_test_split

y = df['Outcome']
X = df.drop('Outcome',axis = 1)

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2) #test_size=1/3,random_state=42, stratify=y
print(len(X_train), len(X_test))
```

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

# Running a machine learning pipeline
def run_ml_pipeline(X_train: pd.DataFrame , y_train: pd.DataFrame, X_test: pd.DataFrame, y_test: pd.DataFrame):
    models = [
              ('Decision tree', DecisionTreeClassifier(), {
                  'max_depth' : [2, 8, 16, 32],
                  'criterion' : ["gini", "entropy"],
                  'max_features': ["auto", "sqrt", "log2"]
              }), 
              ('Gradient Boosting', GradientBoostingClassifier(), {
                  'learning_rate' : [0.1],
                  'n_estimators' : [10, 100, 500],
                  'max_depth' : [2, 8, 16, 32],
                  'max_features': ["auto", "sqrt", "log2"]
              }),
              ('KNN', KNeighborsClassifier(), {
                  'n_neighbors' : [ i for i in range(1,20)]
              }),
              
              ('Naive Bayes', GaussianNB()  , {

              })
            ]
    
    
    results = []
    names = []
    dfs = []
    target_names = ['Outcome']
    scoring = ['accuracy', 'precision_weighted', 'recall_weighted', 'f1_weighted', 'roc_auc']

    for name, model,spec in models:
        kfold = model_selection.KFold(n_splits=5, shuffle=True, random_state=90210)
        search = GridSearchCV(model, spec, scoring='precision', n_jobs=-1, cv=kfold)
        clf = search.fit(X_train, y_train)
        best_model = clf.best_estimator_
        y_pred = best_model.predict(X_test)
        cv_results = model_selection.cross_validate(clf, X_train, y_train, cv=kfold, scoring=scoring)
        print(name)
        print(classification_report(y_test, y_pred))
        results.append(cv_results)
        names.append(name)

        this_df = pd.DataFrame(cv_results)
        this_df['model'] = name
        dfs.append(this_df)
    final = pd.concat(dfs, ignore_index=True)
    return final
```

```python
final = run_ml_pipeline(X_train,y_train, X_test, y_test) # Running a machine learning pipeline
```

```python
bootstraps = []
for model in list(set(final.model.values)):
    model_df = final.loc[final.model == model]
    bootstrap = model_df.sample(n=30, replace=True)
    bootstraps.append(bootstrap)
        
bootstrap_df = pd.concat(bootstraps, ignore_index=True)
results_long = pd.melt(bootstrap_df,id_vars=['model'],var_name='metrics', value_name='values')
time_metrics = ['fit_time','score_time'] # fit time metrics

# Performance metrics
results_long_nofit = results_long.loc[~results_long['metrics'].isin(time_metrics)] # get df without fit data
results_long_nofit = results_long_nofit.sort_values(by='values')

# Time metrics
results_long_fit = results_long.loc[results_long['metrics'].isin(time_metrics)] # df with fit data
results_long_fit = results_long_fit.sort_values(by='values')
```

```python
plt.figure(figsize=(20, 12))
sns.set(font_scale=2.5)
g = sns.boxplot(x="model", y="values", hue="metrics", data=results_long_nofit, palette="Set3")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Comparison of Model by Classification Metric')
plt.savefig('./benchmark_models_performance.png',dpi=300)
```
